# 3. Strings
A string in Python is a sequence of characters enclosed within single quotes (' ') or double quotes (" "). Strings are used to represent text.

```python
# Declaring Strings
single_quote_string = 'Hello, World!'
double_quote_string = "Hello, World!"
```

### Basic String Operation

```python
# 1. Concatenation:Joining two or more strings using the + operator.
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)
# Output: John Doe
```

```python
# 2. Repetition:Repeating a string multiple times using the * operator.
greeting = "Hello! " * 3
print(greeting)
# Output: Hello! Hello! Hello!
```

```python
# 3.0 Indexing: Accessing individual characters of a string using index positions (starting from 0).
text = "Python"
first_char = text[0]
last_char = text[-1]
print(first_char, last_char)
# Output: P n
```

```python
# 4.0 Slicing: Extracting a substring using a range of indices.
text = "Python Programming"
substring = text[0:6]
print(substring)
# Output: Python
```

```python
# 5.0 Length: Finding the length of a string using the len() function.
text = "Hello, World!"
length = len(text)
print(length)
# Output: 13
```

```python
# 6.0 Other string methods to manipulate strings.
# Upper and Lower Case:
text = "Hello, World!"
print(text.upper())
# Output: HELLO, WORLD!
print(text.lower())
# Output: hello, world!
```

```python
# 7.0 Strip:Removing leading and trailing whitespaces.
text = "   Hello, World!   "
stripped_text = text.strip()
print(stripped_text)
# Output: Hello, World!
```

```python
# 8.0 Replace: Replacing a substring with another substring.
text = "Hello, World!"
new_text = text.replace("World", "Python")
print(new_text)
# Output: Hello, Python!
```

```python
# 9.0 Split: Splitting a string into a list of substrings based on a delimiter.
text = "Hello, World!"
words = text.split(", ")
print(words)
# Output: ['Hello', 'World!']
```

```python
# 10. Join:Joining a list of strings into a single string with a delimiter.
words = ["Hello", "World!"]
sentence = " ".join(words)
print(sentence)
# Output: Hello World!
```

```python
# 11. User Input and Greeting
name = input("Enter your name: ")
greeting = "Hello, " + name + "!"
print(greeting)
```

```python
# 12.  Counting Words in a Sentence
sentence = "This is a sample sentence."
words = sentence.split()
word_count = len(words)
print("The number of words in the sentence is:", word_count)
```
